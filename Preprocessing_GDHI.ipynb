```python
#Opening GDHI and population files
GDHI = pd.read_excel('regionalgrossdisposablehouseholdincomelocalauthorities2023.xlsx', sheet_name='Table 3',  header=1)
population = pd.read_excel('regionalgrossdisposablehouseholdincomelocalauthorities2023.xlsx', sheet_name='Table 2',  header=1)
```

FIltering GDHI and Population data to include only the districts/regions we are interested in.

```python
#Creating a list all the districts
region =[
    # Bedfordshire
    'Bedford',
    'Central Bedfordshire',
    'Luton',
    # Hertfordshire
    'Broxbourne',
    'Dacorum',
    'East Hertfordshire',
    'Hertsmere',
    'North Hertfordshire',
    'St Albans',
    'Stevenage',
    'Three Rivers',
    'Watford',
    'Welwyn Hatfield',
    # Surrey
    'Elmbridge',
    'Epsom and EWELL',
    'Guildford',
    'Mole Valley',
    'Reigate and Banstead',
    'Runnymede',
    'Spelthorne',
    'Surrey Heath',
    'Tandridge',
    'Waverley',
    'Woking',
    # Kent
    'Ashford',
    'Canterbury',
    'Dartford',
    'Dover',
    'Folkestone and Hythe',
    'Gravesham',
    'Maidstone',
    'Medway',
    'Sevenoaks',
    'Swale',
    'Thanet',
    'Tonbridge and Malling',
    'Tunbridge Wells'
]

region_filter = '|'.join(region)
```

```python

GDHI_filtered = GDHI[GDHI['Region name'].str.contains(region_filter, case=False, na=False)] #st.contains lets us search for the mention of a particular district in Region name.
population_filtered = population[population['Region name'].str.contains(region_filter, case=False, na=False)]
```

### Adding columns for County 
In the block of code below, the county each row pertains to in the GDHI & population dataset is assigned.

```python
#Creating seperate lists for districts/regions that belong in each County

Bed = [
    'Bedford',
    'Central Bedfordshire',
    'Luton'
]

Surrey =[
    'Elmbridge',
    'Epsom and Ewell',
    'Guildford',
    'Mole Valley',
    'Reigate and Banstead',
    'Runnymede',
    'Spelthorne',
    'Surrey Heath',
    'Tandridge',
    'Waverley',
    'Woking'
]

Kent =[
    'Ashford',
    'Canterbury',
    'Dartford',
    'Dover',
    'Folkestone and Hythe',
    'Gravesham',
    'Maidstone',
    'Medway',
    'Sevenoaks',
    'Swale',
    'Thanet',
    'Tonbridge and Malling',
    'Tunbridge Wells'
]

Hert = ['Broxbourne', 'Dacorum', 'East Hertfordshire','Hertsmere','North Hertfordshire',
                'St Albans','Stevenage','Three Rivers','Watford', 'Welwyn Hatfield']


#  Using 'isin' to assign a 'True' output on rows which contain the name of a region/district that is in the county list
Bed_filter = GDHI_filtered['Region name'].isin(Bed)
Hert_filter = GDHI_filtered['Region name'].isin(Hert)
Surrey_filter = GDHI_filtered['Region name'].isin(Surrey)
Kent_filter = GDHI_filtered['Region name'].isin(Kent)

# Using .loc[] to assign the county name on a new column ('County') whereever there is a true output on the filter
GDHI_filtered.loc[Bed_filter, 'County'] = 'Bedfordshire'
GDHI_filtered.loc[Hert_filter, 'County'] = 'Hertfordshire'
GDHI_filtered.loc[Surrey_filter, 'County'] = 'Surrey'
GDHI_filtered.loc[Kent_filter, 'County'] = 'Kent'

#below is the same code but applied to the population dataset
Bed_filter = population_filtered['Region name'].isin(Bed)
Hert_filter = population_filtered['Region name'].isin(Hert)
Surrey_filter = population_filtered['Region name'].isin(Surrey)
Kent_filter = population_filtered['Region name'].isin(Kent)

population_filtered.loc[Bed_filter, 'County'] = 'Bedfordshire'
population_filtered.loc[Hert_filter, 'County'] = 'Hertfordshire'
population_filtered.loc[Surrey_filter, 'County'] = 'Surrey'
population_filtered.loc[Kent_filter, 'County'] = 'Kent'

#removing any nulls from both datasets
population_filtered.replace(['?', 'NULL', 'NA', ''], np.nan, inplace=True)
GDHI_filtered.replace(['?', 'NULL', 'NA', ''], np.nan, inplace=True)

population_filtered.dropna(subset=['County'],inplace=True)
GDHI_filtered.dropna(subset=['County'],inplace=True)
```

Creating output files for the newly filtered data

```python
GDHI_filtered.to_excel('GDHI.xlsx')
population_filtered.to_excel('population.xlsx')
```
